```python
import pandas as pd
```

```python
# Create sqlalchemy
from sqlalchemy import create_engine
engine = create_engine(f"sqlite:///spotify.sqlite")
```

```python
# Connect to database
with engine.connect() as conn:
    songsdf = pd.read_sql("SELECT * FROM songs", conn)
    tracksdf = pd.read_sql("SELECT * FROM trackinfo", conn)
```

```python
joined_df = pd.merge(songsdf, tracksdf, on='track_id', how='inner')
joined_df.head()
```

```python
joined_df.to_csv("joined_data_tableau.csv", index = False)
```
